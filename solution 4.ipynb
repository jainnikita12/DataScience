```python
def day(x):
    if x == 16 or x == 19:
        x = "16and19"
    elif x == 17 or x == 18:
        x = "17and18"
    elif x == 7 or x == 8:
        x = "7and8"
    elif x == 12 or x == 13:
        x = "12and13"
    elif x == 14 or x == 15:
        x = "14and15"
    elif x == 9:
        x = "9"
    elif x == 0:
        x = "0"
    else:
        x = "other_time"
    return x
```

```python
# dummies for encoding
def create_dummies(var):
    var_unique = var.unique()
    var_unique.sort()
    dummy = pandas.DataFrame()
    for val in var_unique:
    #for val in var_unique[:-1]:
        # which columns are equal to our unique value
        d = var == val
        # make a new column with a dummy variable
        dummy[var.name + "_" + str(val)] = d.astype(int)
    return(dummy)

data['sex_upon_outcome'] =data['sex_upon_outcome'].fillna("Neutered Male")
sex_dummies = create_dummies(data["sex_upon_outcome"])
data = pandas.concat([data, sex_dummies], axis=1)
del data["sex_upon_outcome"]

test['sex_upon_outcome'] =test['sex_upon_outcome'].fillna("Neutered Male")
sex_dummies_test = create_dummies(test["sex_upon_outcome"])
test = pandas.concat([test, sex_dummies_test], axis=1)
del test["sex_upon_outcome"]

outcome_hour_dummies = create_dummies(data["outcome_hour"])
data = pandas.concat([data, outcome_hour_dummies], axis=1)
del data["outcome_hour"]

outcome_hour_dummies_test = create_dummies(test["outcome_hour"])
test = pandas.concat([test, outcome_hour_dummies_test], axis=1)
del test["outcome_hour"]


data.head()
```
